# **Objective**

Financial news headlines are a rich source of NLP data, particularly for predicting stock market performance. A common approach to this task is **sentiment analysis**, which classifies text into positive, negative, or neutral sentiment categories.

In this project, the dataset consists of the top 25 financial market news headlines for the day. The objective is to train a model that can predict the overall sentiment of these headlines, helping to assess market mood and possibly guide investment decisions.

# **Text conversion to bag of words**

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(lowercase = True, ngram_range = (1,1))
```

```python
X = cv.fit_transform(X)
```

```python
X.shape
```

```python
y = df['Label']
```

```python
y.shape
```

# **Train Test Split dataset**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, stratify = y, random_state = 2520)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators = 200)
```

```python
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

# **Explanation**

This result is based on a binary classification task, where there are two classes: 0 and 1.
 Let's break down the key metrics:

This result is based on a binary classification task, where the model predicts one of two classes: 0 or 1. Let’s break down the key performance metrics:

**Precision:**

Precision measures how many of the predicted positives were actually correct. For class 0, the precision is 0.46, meaning 46% of the instances predicted as class 0 were indeed class 0. For class 1, the precision is 0.52, indicating 52% of the instances predicted as class 1 were true positives.

**Recall (Sensitivity):**

Recall, or True Positive Rate, measures how well the model identifies actual positives. For class 0, the recall is 0.22, indicating that only 22% of the true class 0 instances were correctly predicted. For class 1, recall is higher at 0.77, meaning 77% of the true class 1 instances were correctly identified.

**F1-Score:**

The F1-score is the harmonic mean of precision and recall, offering a balanced metric when you want to consider both false positives and false negatives. For class 0, the F1-score is 0.29, and for class 1, it’s 0.62. A higher F1-score indicates a better balance between precision and recall.

**Support:**

Support reflects the number of instances in each class. Here, there are 581 instances of class 0 and 650 instances of class 1 in the dataset.

**Accuracy:**

Accuracy measures the overall correctness of the model by calculating the proportion of correct predictions out of all predictions. In this case, the model’s overall accuracy is 51%.

**Macro Average & Weighted Average:**

- The **macro average** is the unweighted average of precision, recall, and F1-score across both classes, treating them equally.
- The **weighted average** takes into account the number of instances in each class, giving more weight to classes with higher support. This provides a more representative performance measure when class distribution is imbalanced.
